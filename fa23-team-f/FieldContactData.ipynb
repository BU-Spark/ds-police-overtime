```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

field2020 = pd.read_csv('./data/Demographic/fio_2020.csv')

cols_to_drop = ['fc_num', 'vehicle_year', 'vehicle_state', 'vehicle_color', 'vehicle_model', 'vehicle_style', 'vehicle_type', 'weather']

for col in cols_to_drop:
    field2020.drop(col, axis=1, inplace=True)

zipcodes = {}
for code in field2020['zip']:
    if code in zipcodes:
        zipcodes[code] += 1
    else:
        zipcodes[code] = 1
print(zipcodes)
topop = []
for x in zipcodes:
    if zipcodes[x] < 10:
        topop.append(x)
for x in topop:
    zipcodes.pop(x)


plt.bar(zipcodes.keys(), zipcodes.values(), color='g')
plt.xlabel('Zip Code')
plt.ylabel('Number of Reports')
plt.title('Quantity of Reports by Zip Code')
plt.show()
```

```python
with open('./data/Demographic/racesex22.csv') as fd:
    reader22=csv.reader(fd)
    interestingrows22=[row for idx, row in enumerate(reader22) if idx in (3,4,5,7,8,10,11,12,13,14,15)]

x = ['W/M', 'W/F', 'B/M', 'B/F', 'H/M', 'H/F', 'A/M', 'A/F']
superintendentvals22 = [int(x) for x in interestingrows22[0][2:]]
deputyvals22 = [int(x) for x in interestingrows22[1][2:]]
captainvals22 = [int(x) for x in interestingrows22[2][2:]]
captaindetectivevals22 = [int(x) for x in interestingrows22[3][2:]]
lieutenantvals22 = [int(x) for x in interestingrows22[4][2:]]
lieutenantdetectivevals22 = [int(x) for x in interestingrows22[5][2:]]
sargeantvals22 = [int(x) for x in interestingrows22[6][2:]]
sargeantdetectivevals22 = [int(x) for x in interestingrows22[7][2:]]
detectivevals22 = [int(x) for x in interestingrows22[8][2:]]
patrolofficervals22 = [int(x) for x in interestingrows22[9][2:]]
totalsworn22 = [str.replace(',', '') for str in interestingrows22[10][2:]]
totalsworn22 = [int(x) for x in totalsworn22]

with open('./data/Demographic/racesex23.csv') as fdd:
    reader23=csv.reader(fdd)
    interestingrows23=[row for idx, row in enumerate(reader23) if idx in (3,4,5,7,8,10,11,12,13,14,15)]

superintendentvals23 = [int(x) for x in interestingrows23[0][2:]]
deputyvals23 = [int(x) for x in interestingrows23[1][2:]]
captainvals23 = [int(x) for x in interestingrows23[2][2:]]
captaindetectivevals23 = [int(x) for x in interestingrows23[3][2:]]
lieutenantvals23 = [int(x) for x in interestingrows23[4][2:]]
lieutenantdetectivevals23 = [int(x) for x in interestingrows23[5][2:]]
sargeantvals23 = [int(x) for x in interestingrows23[6][2:]]
sargeantdetectivevals23 = [int(x) for x in interestingrows23[7][2:]]
detectivevals23 = [int(x) for x in interestingrows23[8][2:]]
patrolofficervals23 = [int(x) for x in interestingrows23[9][2:]]
totalsworn23 = [str.replace(',', '') for str in interestingrows23[10][2:]]
totalsworn23 = [int(x) for x in totalsworn23]
```
