```python
import matplotlib as plt
import csv
import pandas as pd
import numpy as np

field = pd.read_csv('2023fielddata.csv')
field.head()
```

```python
def extract_zips(data, dict):
    if type(data) != int:
        return
    else:
        strdata = str(data)
        if len(strdata) == 4:
                strdata = '0' + strdata
        if strdata not in dict:
            dict[strdata] = 1
        else:
            dict[strdata] += 1
```

```python
zips = field['zip']
zips.head()
zipdict = {}
for row in field['zip']:
    extract_zips(row, zipdict)
print(zipdict)
```
