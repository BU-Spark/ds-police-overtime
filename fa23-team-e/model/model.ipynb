```python
def contains_alphabetic(input_string):
    return any(char.isalpha() for char in input_string)

def convert_data(value):
    if value == None:
      return 0.0
    elif not isinstance(value, float):
      if '-' in value:
          return np.nan
      elif '(' in value:
         return -1 * float(value.replace("$", "").replace(",", "").replace("(", "").replace(")", ""))
      elif not contains_alphabetic(value):
          return float(value.replace("$", "").replace(",", ""))
      else:
          return 0.0
      
      
monetary_columns = ['REGULAR', 'RETRO', 'OTHER', 'OVERTIME', 'INJURED', 'DETAIL', 'EDUCATION', 'TOTAL_EARNING']

# Function to convert monetary values to floats
# def convert_to_float(monetary_value):
#     # Remove currency symbols and commas, and convert parentheses to negative numbers
#     if isinstance(monetary_value, str):
#         clean_value = monetary_value.replace('$', '').replace(',', '')
#         if '(' in clean_value and ')' in clean_value:
#             clean_value = clean_value.replace('(', '').replace(')', '')
#             return float(clean_value) * -1
#         elif '  -  ' in clean_value:
#             return 0.0
#         else:
#             return float(clean_value)
#     else:
#         return monetary_value

for df in earning_data_list:
  for column in monetary_columns:
      df[column] = df[column].apply(convert_data)
```

```python
import re

def clean_text(text):
    # Replace non-alphabetic characters (including commas, periods, numbers, etc.) with an empty string
    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text)
    return cleaned_text

# Categorize the crimes  
def categorize_offense(offense):

    offense = str(offense).lower()


    violent_crimes = [
        'Assault', 'Aggravated Assault', 'Simple Assault', 'Homicide', 'Firearm Violations', 
        'Ballistics', 'HOME INVASION', 'Offenses Against Child / Family', 'Harassment', 
        'Criminal Harassment', 'Threats to Do Bodily Harm', 'Murder, Non-Negligient Manslaughter', 
        'Manslaughter', 'Homicide', 'Manslaughter - Negligence', 'Manslaughter - Vehicle - Negligence',
        'Justifiable Homicide', 'Assault - Aggravated', 'Assault - Simple', 'Assault Simple - Battery', 
        'Assault - Aggravated - Battery', 'Harassment/ Criminal Harassment'
    ]
    
    robbery = [
        'Robbery', 'Robbery - Street', 'Robbery - Other', 'Robbery - Home Invasion', 'Robbery - Commercial', 
        'Robbery - Bank', 'Robbery - Car Jacking'
    ]

    aggravated_assault = [
        'Aggravated Assault', 'Assault - Aggravated', 'Assault - Aggravated - Battery'
    ]

    property_crimes = [
        'Fraud', 'Property Lost', 'Property Found', 'Vandalism', 'Towed', 'Property Related Damage', 
        'Recovered Stolen Property', 'Confidence Games', 'Embezzlement', 'Counterfeiting', 'Forgery / Counterfeiting',
        'Fraud - False Pretense / Scheme', 'Fraud - Impersonation', 'Fraud - Credit Card / ATM Fraud', 
        'Fraud - Welfare', 'Fraud - Wire', 'Property - Lost/ Missing', 'Property - Accidental Damage', 
        'Property - Lost Then Located', 'Property - Stolen Then Recovered', 'Property - Concealing Leased'
    ]

    burglary = [
        'Residential Burglary', 'Commercial Burglary', 'Other Burglary', 'Burglary', 'Burglary - No Property Taken',
        'Burglary - Residential', 'Burglary - Commerical', 'Burglary - Other', 'Burglary - Other - Force', 
        'Burglary - Other - No Force', 'Burglary - Commerical - Force', 'Burglary - Commerical - No Force', 
        'Burglary - Residential - Force', 'Burglary - Residential - No Force', 'Burglary - Residential - Attempt', 
        'Burglary - Commerical - Attempt', 'Burglary - Other - Attempt'
    ]

    larceny = [
        'Larceny', 'Larceny From Motor Vehicle', 'Auto Theft', 'Auto Theft Recovery', 'Larceny All Others',
        'Larceny Theft From MV - Non-Accessory', 'Larceny Theft of MV Parts & Accessories', 
        'Larceny Theft from Building', 'Larceny Shoplifting', 'Larceny Pick-Pocket', 'Larceny Purse Snatch - No Force', 
        'Larceny Theft from Coin-Op Machine', 'Larceny Theft of Bicycle'
    ]

    vehicle_related = [
        'Motor Vehicle Accident Response', 'Operating Under the Influence', 'License Plate Related Incidents', 
        'Aircraft', 'M/V Accident - Other', 'M/V Accident - Personal Injury', 'M/V Accident - Property Damage', 
        'M/V Accident - Involving Pedestrian - Injury', 'M/V Accident - Involving Bicycle - Injury', 
        'M/V - Leaving Scene - Personal Injury', 'M/V - Leaving Scene - Property Damage', 
        'M/V Accident - Other City Vehicle', 'M/V Accident - Police Vehicle', 'M/V Accident - Invovling Pedestrian - No Injury',
        'M/V Accident - Involving Bicycle - No Injury', 'M/V Accident - Invovling Bicycle - Injury', 
        'Auto Theft - Motorcycle / Scooter', 'Auto Theft - Leased/Rented Vehicle'
    ]

    if offense in violent_crimes or 'assualt' in offense or 'homicide' in offense or 'firearm' in offense or 'home invasion' in offense or 'harassment' in offense or 'threats' in offense or 'murder' in offense or 'manslaughter' in offense or 'killing' in offense or 'threat' in offense:
        return 'violent'
    elif offense in robbery or 'robbery' in offense:
        return 'robbery'
    elif offense in aggravated_assault or 'assault' in offense or 'aggravated' in offense:
        return 'aggravated assault'
    elif 'fraud' in offense or offense in property_crimes or 'property' in offense or 'vandalism' in offense or 'embezzlement' in offense or 'counterfeiting' in offense or 'forgery' in offense or 'confidence games' in offense:
        return 'property'
    elif offense in burglary or 'burglary' in offense:
        return 'burglary'
    elif offense in larceny or 'larceny' in offense or 'theft' in offense:
        return 'larceny'
    elif offense in vehicle_related or 'm/v' in offense or 'auto' in offense or 'motor' in offense or 'operating' in offense or 'driving' in offense or 'vehicle' in offense:
        return 'vehicle'
    else:
        # with open('other_crimes.txt', 'a') as file:
        #     file.write(f"{offense} \n")
        return 'other'



for df in c_i_list:
    df['Category'] = None
    # Apply categorize_offense to the first five rows based on 'OFFENSE_CODE_GROUP'
    df.iloc[:4, df.columns.get_loc('Category')] = df.iloc[:4]['OFFENSE_CODE_GROUP'].apply(categorize_offense)

    # Apply categorize_offense to the rest of the rows based on 'OFFENSE_DESCRIPTION'
    df.iloc[4:, df.columns.get_loc('Category')] = df.iloc[4:]['OFFENSE_DESCRIPTION'].apply(categorize_offense)



```
